```python
#sin波形(模擬訊號偏強的圖 頻率=2/3)
import numpy as np
from matplotlib import pyplot as plt

for i in range(0,3,1):
  x=np.arange(0,45*np.pi,0.0415)
  #print(x.shape)
  y=np.sin(x)
  #print(y.shape)
  if i==0:
    zero_set=np.zeros([0])#無平移
    print('▼模擬sin波的數據,推遲0秒▼')
  elif i==1:
    zero_set=np.zeros([100])#平移0~1秒
    print('▼模擬sin波的數據,推遲1秒▼')
  elif i==2:
    zero_set=np.zeros([200])#平移0~2秒  
    print('▼模擬sin波的數據,推遲2秒▼')

  y = np.insert(y,0,zero_set)

  y_sin3000point = y[0:3000]
  #print(y_sin3000point.shape)
  
  y_sin1x3000point = y_sin3000point.reshape(1,3000)
  #print(y_sin1x3000point.T.shape)
  plt.axis('on')
  plt.plot(y_sin1x3000point.T)#確認(3000,)轉成(3000,1)是否有問題,波形圖
  plt.show()
  #改壓縮成10x300
  y_sin10x300point = y_sin3000point.reshape(10,300)
  #print(y_sin10x300point.shape)
  #print(y_sin10x300point[0].shape)
  print('▼模擬sin波的數據,將上面數據切成10份,不要第1份,共9份數據▼')
  for j in range(1,10,1):
    plt.subplot(1,9,j)
    axx=plt.gca()
    axx.get_xaxis().set_visible(False)
    axx.get_yaxis().set_visible(False)
    plt.plot(y_sin10x300point[j].T)#確認(3000,)轉成(3000,1)是否有問題,波形圖
  plt.show()
  print('▼將上面數據增至15通道,1人9份數據▼')
  for j in range(1,10,1):
    data = y_sin10x300point[j]
    #print('data長度=',data.shape)
    for a in range(0,14,1):#這裡決定數據有幾通道,目前預設14+1=15通道數據
      data = np.vstack([data,y_sin10x300point[j]])
    
    #plt.figure(figsize=(77/300, 62/300), dpi=300)
    data75x60 = data.reshape(75,60)
    
    
    #plt.imshow(data75x60)
    zero_set=np.zeros([77,62])
    data77x62=np.pad(data75x60,1)
    
    plt.subplot(1,9,j)
    plt.axis('off')
    plt.imshow(data77x62,cmap='Greys')

    #SavePath = ('gdrive/MyDrive/南台專案工作筆記/類神經網路CNN/深度學習/20220621N值CNN驗證/train/Lv1.%d_%d.png' %(i,j))
    #plt.savefig(SavePath, transparent=True, pad_inches=0)
  plt.show()
```

# CNN方法(暫時只有2值)

```python
#訓練集路徑確認
!ls gdrive/MyDrive/南台專案工作筆記/類神經網路CNN/深度學習/20220621N值CNN驗證/train
```

```python
import numpy as np 
import pandas as pd 
import cv2
import matplotlib.pyplot as plt
import tensorflow as tf
import os
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, Activation, Conv2D, MaxPooling2D

#依名稱設定成:Lv1=0,Lv2=1做訓練

train_dir = 'gdrive/MyDrive/南台專案工作筆記/類神經網路CNN/深度學習/20220621N值CNN驗證/train'
path = os.path.join(train_dir)
X = []
y = []
convert = lambda category : int(category == 'Lv2')
def create_test_data(path):
    for p in os.listdir(path):
        category = p.split(".")[0]
        category = convert(category)
        img_array = cv2.imread(os.path.join(path,p),cv2.IMREAD_GRAYSCALE)
        new_img_array = cv2.resize(img_array, dsize=(80, 80))
        X.append(new_img_array)
        y.append(category)
create_test_data(path)
X = np.array(X).reshape(-1, 80,80,1)
y = np.array(y)
#Normalize data
X = X/255.0

#卷積設定
model = Sequential()
# Adds a densely-connected layer with 10 units to the model:
model.add(Conv2D(10,(2,2), activation = 'relu', input_shape = X.shape[1:]))

model.add(MaxPooling2D(pool_size = (2,2)))
# Add another:
model.add(Conv2D(15,(3,3), activation = None))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Flatten())
model.add(Dense(20, activation='relu'))
# Add a softmax layer with 10 output units:
model.add(Dense(1, activation='sigmoid'))
model.summary()
model.compile(optimizer="adam",
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

## CNN訓練

```python
#訓練20次
model.fit(X, y, epochs=20, batch_size=32)
```
