# Sin波不同頻率用來測試CNN的N值判斷是否能夠正確判別。

**此Sin波模擬類似上肢啞鈴曲舉的前臂角度變化或加速度變化，用於分析六軸感測系統IMU之分析方法。**

Lv1-正常,訓練量3張

Lv2-正常但漸弱,訓練量12張

Lv3-亞健康,訓練量3張

Lv4-衰弱(亞健康+漸弱),訓練量12張

Lv5-中途放棄(頻率1.5Hz),訓練量24張

Lv6-中途放棄(頻率3Hz),訓練量24張

## 頻率1.5Hz組(健康~亞健康)

```python
#sin波形(模擬訊號偏強的圖 頻率=1.5Hz)
import numpy as np
from matplotlib import pyplot as plt

Lv=1

for i in range(0,12,1):
  x=np.arange(0,45*np.pi,0.0415)
  #print(x.shape)
  y=np.sin(x)
  #print(y.shape)
  if i==0:
    zero_set=np.zeros([0])#無平移
    print('▼模擬sin波的數據,推遲0秒,訓練成:Lv1(正常)▼')
    y = np.insert(y,0,zero_set)
    y_sin3000point = y[0:3000]
    Lv=1
  elif i==1:
    zero_set=np.zeros([100])#平移1秒
    print('▼模擬sin波的數據,推遲1秒,訓練成:Lv1(正常)▼')
    y = np.insert(y,0,zero_set)
    y_sin3000point = y[0:3000]
    Lv=1
  elif i==2:
    zero_set=np.zeros([200])#平移2秒  
    print('▼模擬sin波的數據,推遲2秒,訓練成:Lv1(正常)▼')
    y = np.insert(y,0,zero_set)
    y_sin3000point = y[0:3000]
    Lv=1
    
  elif i==3: 
    zero_set=np.zeros([0])#無平移
    print('▼模擬sin波的數據,斜率拉升,訓練成:Lv2(漸弱)▼')
    Yevent = np.arange(0,0.3407,0.0001)
    #print(Yevent.shape)
    y = np.add(y, Yevent)
    y_sin3000point = y[0:3000]
    Lv=2
  elif i==4: 
    zero_set=np.zeros([0])#無平移
    print('▼模擬sin波的數據,斜率拉降,訓練成:Lv2(漸弱)▼')
    Yevent = np.arange(0,0.3407,0.0001)
    #print(Yevent.shape)
    y = y - Yevent
    y_sin3000point = y[0:3000]
    Lv=2
  elif i==5: 
    zero_set=np.zeros([0])#無平移
    print('▼模擬sin波的數據,斜率收縮,訓練成:Lv2(漸弱)▼')
    Yevent = np.arange(0,102.21,0.0001)
    #print(Yevent.shape)
    Yevent = Yevent[6814:10221]
    #print(Yevent.shape)
    y = y * Yevent 
    y_sin3000point = y[3000:0:-1]
    Lv=2
  elif i==6: 
    zero_set=np.zeros([100])#平移1秒
    print('▼模擬sin波的數據,斜率拉升,訓練成:Lv2(漸弱)▼')
    Yevent = np.arange(0,0.3407,0.0001)
    #print(Yevent.shape)
    y = np.add(y, Yevent)
    y = np.insert(y,0,zero_set)
    y_sin3000point = y[0:3000]
    Lv=2
  elif i==7: 
    zero_set=np.zeros([100])#平移1秒
    print('▼模擬sin波的數據,斜率拉降,訓練成:Lv2(漸弱)▼')
    Yevent = np.arange(0,0.3407,0.0001)
    #print(Yevent.shape)
    y = y - Yevent
    y = np.insert(y,0,zero_set)
    y_sin3000point = y[0:3000]
    Lv=2
  elif i==8: 
    zero_set=np.zeros([100])#平移1秒
    print('▼模擬sin波的數據,斜率收縮,訓練成:Lv2(漸弱)▼')
    Yevent = np.arange(0,102.21,0.0001)
    #print(Yevent.shape)
    Yevent = Yevent[6814:10221]
    #print(Yevent.shape)
    y = y * Yevent 
    y_sin3000point = y[3000:0:-1]
    y_sin3000point = np.insert(y_sin3000point,0,zero_set)
    Lv=2
  elif i==9: 
    zero_set=np.zeros([200])#平移2秒 
    print('▼模擬sin波的數據,斜率拉升,訓練成:Lv2(漸弱)▼')
    Yevent = np.arange(0,0.3407,0.0001)
    #print(Yevent.shape)
    y = np.add(y, Yevent)
    y = np.insert(y,0,zero_set)
    y_sin3000point = y[0:3000]
    Lv=2
  elif i==10: 
    zero_set=np.zeros([200])#平移2秒 
    print('▼模擬sin波的數據,斜率拉降,訓練成:Lv2(漸弱)▼')
    Yevent = np.arange(0,0.3407,0.0001)
    #print(Yevent.shape)
    y = y - Yevent
    y = np.insert(y,0,zero_set)
    y_sin3000point = y[0:3000]
    Lv=2
  elif i==11: 
    zero_set=np.zeros([200])#平移2秒 
    print('▼模擬sin波的數據,斜率收縮,訓練成:Lv2(漸弱)▼')
    Yevent = np.arange(0,102.21,0.0001)
    #print(Yevent.shape)
    Yevent = Yevent[6814:10221]
    #print(Yevent.shape)
    y = y * Yevent 
    y_sin3000point = y[3000:0:-1]
    y_sin3000point = np.insert(y_sin3000point,0,zero_set)
    Lv=2
    
  #print(y_sin3000point.shape)
  #扣除前300點(開始3秒因延遲與人為反應問題,所以不計算)
  y_sin2700point = y_sin3000point[300:3000]
  #print(y_sin2700point.shape)
  #調整維度1:2700
  y_sin2700point = y_sin2700point.reshape(1,2700)
  #print(y_sin1x2700point.T.shape)
  plt.axis('on')
  plt.plot(y_sin2700point.T)#確認(3000,)轉成(3000,1)是否有問題,波形圖
  plt.show()

  #調整維度成45:60並新開一個變數儲存
  y_sin45x60point = y_sin2700point.reshape(45,60)
  
  #加白色外框避免失真
  zero_set=np.zeros([47,62])
  data47x62=np.pad(y_sin45x60point,1)
  data = zero_set+data47x62

  #輸出灰階圖
  plt.axis('off')
  plt.imshow(data,cmap='Greys')
#  SavePath = ('gdrive/MyDrive/南台專案工作筆記/類神經網路CNN/深度學習/20220623N值CNN驗證/train/Lv%d.%d.png' %(Lv,i))
#  plt.savefig(SavePath, transparent=True, pad_inches=0)
  plt.show()
  
```

```python
#讀圖測試
import numpy as np
import cv2
import os

train_dir = 'gdrive/MyDrive/南台專案工作筆記/類神經網路CNN/深度學習/20220623N值CNN驗證/train'
path = os.path.join(train_dir)
X = []
y = []

def Lv_convert(get_category):#訓練陣列指引
  if get_category=='Lv1':
    return 0
  elif get_category=='Lv2':
    return 1
  elif get_category=='Lv3':
    return 2
  elif get_category=='Lv4':
    return 3
  elif get_category=='Lv5':
    return 4
  elif get_category=='Lv6':
    return 5

def create_train_data(path):#輸出訓練陣列
    for p in os.listdir(path):
        category = p.split(".")[0]
        category = Lv_convert(category)
        img_array = cv2.imread(os.path.join(path,p),cv2.IMREAD_GRAYSCALE)
        new_img_array = cv2.resize(img_array, dsize=(288, 432))
        X.append(new_img_array)
        y.append(category)

create_train_data(path)

X = np.array(X).reshape(-1, 288, 432, 1)
y = np.array(y)
print(X.shape)
print(y.shape)
print(y)
```

```python
from tensorflow import keras
from tensorflow.keras import layers

CNN=keras.Sequential(name='SinWaveCNN')
CNN.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(288,432,1)))#第一層需有input_shape
CNN.add(layers.MaxPooling2D((2,2)))

CNN.add(layers.Conv2D(64,(12,12),activation='relu'))
CNN.add(layers.MaxPooling2D((3,3)))

CNN.add(layers.Flatten())

CNN.add(layers.Dense(128,activation='relu'))
CNN.add(layers.Dense(64,activation='relu'))
CNN.add(layers.Dense(6,activation='softmax'))#0~5(6 layer)

keras.utils.plot_model(CNN,show_shapes=True)
```

```python
CNN.compile(optimizer='Adam',
      loss=keras.losses.sparse_categorical_crossentropy,
      metrics=['accuracy'])
```

```python
#訓練10次
CNN.fit(X, y, epochs=10, batch_size=32)
```
